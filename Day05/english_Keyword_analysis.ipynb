```python
all_data = []
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = pd.concat(all_data, axis=0, ignore_index=True)
all_data_concat
```

```python
all_data_concat.to_csv('./data/riss_bigdata.csv',index=False, encoding='utf-8')
```

```python
all_data_concat=all_data_concat.drop(all_data_concat.columns[[0]],axis='columns')
all_data_concat
```

#### 대이터 전처리

```python
all_title=all_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
stopWords
```

```python
lemma = WordNetLemmatizer()
lemma
```

```python
words = []

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+', ' ', str(title))
    EnWorldsToken = word_tokenize(EnWords.lower())
    EnWorldsTokenStop = [w for w in EnWorldsToken if w not in stopWords]
    EnWorldsTokenStopLemma = [lemma.lemmatize(w) for w in EnWorldsTokenStop]
    words.append(EnWorldsTokenStopLemma)
```

```python
print(words) #불용어 제거 남은 키워드 단어만 남음
```

```python
#2차원 리스트 -> 1차원 리스트로 (차원 축소)
words_reduce = list(reduce(lambda x, y: x+y, words))
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
word_count= dict()
for tag, counts in count.most_common(50):
    if len(str(tag)) > 1:
        word_count[tag]=counts
        #print(f'{tag}:{count}')
word_count
```

#### 단어 대수 차트

```python
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)

plt.figure(figsize=(20,5))
plt.bar(range(len(word_count)),sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='85')

plt.show()
```

#### 워드클라우드

```python
stopwords=set(STOPWORDS)
wc=WordCloud(background_color='ivory', stopwords=stopwords, width=1024, height=700)
cloud=wc.generate_from_frequencies(word_count)

plt.figure(figsize=(15,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
